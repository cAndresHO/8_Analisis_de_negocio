#### ¿Cuál es la duración de cada sesión?

```python
visits['sessions_duration_sec']= (visits['end_ts'] - visits['start_ts']).dt.seconds
print(f"En promedio, la duración de cada sesión fue de: {visits['sessions_duration_sec'].mean()} segundos.")
```

```python
fig5= px.histogram(visits['sessions_duration_sec'],
                  title= 'Distribución de Duración de Cada Sesión en Segundos',
                  nbins=100)
fig5.add_vline(x=visits['sessions_duration_sec'].mean(), line_width=1, line_dash="dash", line_color="red")
fig5.show()
```

**Gráfico 5.** Muestra la distribución de la duración de cada sesión.

Podemos observar (en el *gráf. 5*) que la distribución no es normal. Adicionalmente, existen valores atípicos. Por lo tanto, utilizar la media como indicador no es muy recomendable, en su lugar es mejor utilizar la moda.

```python
print(f"La moda de la duración de cada sesión fue de: {visits['sessions_duration_sec'].mode()}")
```

La mayor cantidad de sesiones dura alrededor de 60 segundos (1 minuto).

```python
first_visits = visits.groupby("uid")['start_ts'].min().reset_index()
```

```python
first_visits.head()
```

```python
first_orders = orders.groupby("uid")['buy_ts'].min().reset_index()
```

```python
first_orders.head()
```

```python
time_to_first_order = first_orders.merge(first_visits, on='uid')
```

```python
time_to_first_order['time_to_first_order'] = time_to_first_order['buy_ts'] - time_to_first_order['start_ts']
```

```python
#Analisis por día
time_to_first_order['time_to_first_order'] = time_to_first_order['time_to_first_order'].dt.days
time_to_first_order.head()
```

```python
print(time_to_first_order['time_to_first_order'].max())
print(time_to_first_order['time_to_first_order'].min())
```

```python
time_to_first_order['time_to_first_order'].describe()
```

```python
time_to_first_order['time_to_first_order'].value_counts().reset_index().rename(columns={'time_to_first_order':'days', 'count': 'time_to_first_order'}).sort_values(by='days')
```

```python
fig6= px.histogram(time_to_first_order['time_to_first_order'],
                   nbins=50,
                   title= 'Cantidad de días desde el primer registro hasta la primera compra',
                   labels={'value': 'Número de días'}
                  )
fig6.add_vline(x=time_to_first_order['time_to_first_order'].mean(), line_width=1, line_dash="dash", line_color="red")
fig6.update_layout(showlegend=False)
fig6.show()
```

**Gráfico 6.** Muestra la cantidad de días que transcurren desde la primera sesión de cada usuario hasta su primera compra.\
>Podemos observar que, aunque la media es de 17 días aproximadamente, la mayoría de los usuarios que incian sesión por primera vez realizan su primera compra en los primeros 10 días.

```python
print(f"La cantidad promedio de días desde la primera sesión hasta la primera compra de cada usuario es de: {time_to_first_order['time_to_first_order'].mean()} días.")
```

```python
print(f"La mediana de la cantidad  de días desde la primera sesión hasta la primera compra de cada usuario es de: {time_to_first_order['time_to_first_order'].median()} días.")
```

```python
print(f"La moda de la cantidad  de días desde la primera sesión hasta la primera compra de cada usuario es de: {time_to_first_order['time_to_first_order'].mode()}")
```
